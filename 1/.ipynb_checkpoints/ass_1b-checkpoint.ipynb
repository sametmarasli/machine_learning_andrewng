## Linear regression with multiple variables

In this part, linear regression with multiple variables is implemented to predict  the  prices  of  houses. Suppose  you  are  selling  your  house  and  you want to know what a good market price would be.  One way to do this is to first collect information on recent houses sold and make a model of housing prices.

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import matplotlib
import os
%matplotlib inline
# Update the matplotlib configuration parameters: 
matplotlib.rcParams.update({'font.size': 12, 'font.family': 'ubuntu'})
```

## Loading Data

```python
directory = os.getcwd()+ '/ex1data2.txt'
data = pd.read_csv(directory, header=None, names= ['size','noBedrooms','price'])

data.describe()
```

## Feature Normalization

```python
def featureNorm(X):   
    '''
    FEATURENORMALIZE(X) returns a normalized version of X where
    the mean value of each feature is 0 and the standard deviation
    is 1. This is often a good preprocessing step to do when
    working with learning algorithms.
    '''
    sigma = np.std(X,axis=0,ddof=1) # ddof=0 population st dev
    mu = np.mean(X,axis=0)
    X = (X-mu)/sigma
    return X,sigma,mu
```

```python
#compare normalized and raw feature values
plt.subplots()
plt.xlim([-100,5000])
plt.hist(data['size'],label = 'col1')
plt.hist(data['noBedrooms'],label = 'col2')
plt.title('Features without normalization')
plt.xlabel('Column Value')
plt.ylabel('Counts')
plt.legend()


plt.subplots()
plt.xlim([-5,5])
plt.hist(featureNorm(data[['size']].values)[0],label = 'col1')
plt.hist(featureNorm(data[['noBedrooms']].values)[0],label = 'col2')
plt.title('Features with normalization')
plt.xlabel('Column Value')
plt.ylabel('Counts')
plt.legend()
```

## Gradient Descent

```python
def computeCostMulti(X,y,theta):
    '''    
    J = COMPUTECOSTMULTI(X, y, theta) computes the cost of using theta as the
    parameter for linear regression to fit the data points in X and y

    '''    
    m,n = X.shape
    theta = theta.reshape(n,1)
    residual = X.dot(theta)-y
    J = 1./(2*m) * residual.T.dot(residual)
    return J

def gradientDescentMulti(X,y,theta,alpha,num_iters):
    '''    
    GRADIENTDESCENTMULTI Performs gradient descent to learn theta
    theta = GRADIENTDESCENTMULTI(x, y, theta, alpha, num_iters) updates theta by
    taking num_iters gradient steps with learning rate alpha

     '''   
    m,n = X.shape
    theta = theta.reshape(n,1)
    J_val = np.zeros((num_iters,1))
    
    for i in range(num_iters):
        residual = np.dot(X,theta)-y
        grad = np.dot(X.T,residual)
        theta = theta - (alpha/m)*grad
        J_val[i] = computeCostMulti(X,y,theta)
    
    return theta,J_val
```

```python
# assing variables
X = data[['size','noBedrooms']].values
m ,n = X.shape
y = data['price'].values.reshape(m,1)

# normalize features
Xnorm,sigma,mu = featureNorm(X)

# add the bias terms
X = np.insert(Xnorm, 0, 1,axis=1)
X.shape

# initialize parameters 
alpha = 0.01
num_iters = 400

# initialize theta
theta = np.zeros(3)
print X.shape, y.shape, theta.shape

# compute theta and J_val's
theta , J_val =  gradientDescentMulti(X,y,theta,alpha,num_iters)

print '\nTheta computed from gradient descent:'
print '{}'.format(theta)
```

```python
# Plot the convergence graph

plt.plot(range(num_iters),J_val)
plt.ylabel('Cost J')
plt.xlabel('Number of Iterations')
plt.title('The Convergence Graph for alpha = {}'.format(alpha))
```

```python
# Estimate the price of a 1650 sq-ft, 3 br house

new_X = np.array([1650.,3.])
price = np.dot(np.append(1,(new_X-mu)/sigma),theta)
print '\nPredicted price of a 1650 sq-ft, 3 br house is ${}'.format(price[0])
```

## Playing with Alpha

```python
# initialize alpha values
alps = np.linspace(0.01,0.1,5)
# np.arange(0.01,0.1,0.02)
num_iters = 50

# Plot J vals for different levels of alpha
fig, ax = plt.subplots(figsize=(10,7))
for e in range(len(alps)):
    ax.plot(range(num_iters),gradientDescentMulti(X,y,theta,alps[e],num_iters)[1],
            label='alpha = {}'.format(alps[e]))
ax.set_ylabel('Cost J')
ax.set_xlabel('Number of Iterations')
ax.set_title('Costs with different alpha levels')
ax.legend(loc=1)
```

## Normal Equations

```python
def normalEqn(X,y):    
    '''theta = pinv(X'X)X'y'''
    theta = np.dot(np.linalg.pinv(X.T.dot(X)),X.T.dot(y)) 
    return theta
```

```python
# prepare data without normalization
X = data[['size','noBedrooms']].values
m ,n = X.shape
y = data['price'].values.reshape(m,1)

# add the bias terms
X = np.insert(X, 0, 1,axis=1)
X.shape

# weights calculated by normal equation
theta_norm = normalEqn(X,y)
print '\nTheta computed from normal equations:'
print '{}'.format(theta_norm)

# Estimate the price of a 1650 sq-ft, 3 br house

new_X = np.array([1,1650,3]).reshape(3,1)
price = np.dot(new_X.T,theta_norm)

print '\nPredicted price of a 1650 sq-ft, 3 br house is {}$'.format(price[0][0])
```
